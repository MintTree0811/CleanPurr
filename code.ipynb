```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Main.csv')
```

## Data Preparation

```python
df.dropna(subset=['Total Sales'],inplace=True)
df.drop(['Total Shipped'],axis =1 ,inplace=True)
```

```python
df.reset_index(inplace=True)
df
```

```python
df.drop(['index','Position'],axis =1 ,inplace=True)
```

```python
df
```

```python
dfm = df[['Game Name','Console','Publisher','Developer','Last Update','NA Sales','PAL Sales','Japan Sales','Other Sales']]
```

```python
output = (dfm.set_index(['Game Name','Console','Publisher','Developer','Last Update'])
   .stack(dropna=False)
   .reset_index()
   .rename({'level_5': 'Region', '0':'Sales'}, axis=1))
```
